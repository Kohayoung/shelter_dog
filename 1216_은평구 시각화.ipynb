```python
#1인가구 거주형태별 남녀 차이
df_j_test = df_eun.iloc[:, 57:69]
df_j_test['단독주택 남-녀']=df_j_test['단독주택_남'] - df_j_test['단독주택_여']
df_j_test['아파트 남-녀']=df_j_test['아파트_남'] - df_j_test['아파트_여']
df_j_test['연립주택 남-녀']=df_j_test['연립주택_남'] - df_j_test['연립주택_여']
df_j_test['다세대주택 남-녀']=df_j_test['다세대주택_남'] - df_j_test['다세대주택_여']
df_j_test['비거주용 건물내 주택 남-녀']=df_j_test['비거주용 건물내 주택_남'] - df_j_test['비거주용 건물내 주택_여']
df_j_test['주택이외의 거처 남-녀']=df_j_test['주택이외의 거처_남'] - df_j_test['주택이외의 거처_남']

df_j_test = df_j_test.iloc[:, 12:19]
df_j_test
```

```python
#1인가구 거주형태별 남녀 차이 [시각화]
ax = df_j_test.plot(kind="line", title="1인가구 거주형태별 남녀 차이", figsize=(12,4), legend=True, fontsize=12)
ax.set_xlabel("년도", fontsize=12)
ax.set_ylabel("차이값", fontsize=12)
ax.legend(['단독주택 남-녀','아파트 남-녀','연립주택 남-녀','다세대주택 남-녀','비거주용 건물내 주택 남-녀','주택이외의 거처 남-녀'], fontsize=12)
```
