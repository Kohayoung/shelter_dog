```python
df05 = pd.read_excel('./데이터/05_서울시 1인가구(연령별)통계[10-18].xls')
df05
```

```python
#컬럼이 너무 많으니 1. 20대 미만, 20대, 30대, 40대, 50대, 60대, 70대 이상 으로 만들고
#2. 각각 남, 여 를 추가해주겠다.
#그럼 총 21개의 컬럼이네
#그럼 1번 먼저
df05['20대'] = df05['20~24세'] + df05['25~29세']
df05
```

```python
#1. 20대 미만, 20대, 30대, 40대, 50대, 60대, 70대 이상 으로 만들기
for i in range(2, 8):
    df05[f'{i}0대']=df05[f'{i}0~{i}4세'] + df05[f'{i}5~{i}9세']
df05
```

```python
#70대는 70대와 80~84세, 85세 이상을 더해줘야 한다.
df05['70대'] = df05['70대']+df05['80~84세']+df05['85세이상']
df05
```
