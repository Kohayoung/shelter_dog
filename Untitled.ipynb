## * 연도별/지역별/인구 수 3d 시각화

```python


df_3d_person = df.iloc[:,[0,1,2]]

#예쁜 시각화를 위해 자치구가 '합계'인 ROW 삭제
df_3d_person = df_3d_person[df_3d_person['자치구'] != '합계']

#연도 컬럼을 인덱싱 -  자치구 값이 '합계'인 ROW 삭제 후 ROW INDEX도 삭제되어 리셋하기 위함 
df_3d_person= df_3d_person.set_index('년도')

#년도로 인덱싱된 데이터 프레임의 인덱스를 정수로 리셋
df_3d_person =df_3d_person.reset_index()
df_3d_person
```

```python
#축을 만들기 위해 연도, 자치구 리스트 뽑기
#연도, 지역 - 컬럼으로 추출
#X = df_year, Y = df_regiond, data = 인구 수
        
df_year = df_3d_person.drop_duplicates('년도', keep='first')['년도']
df_region = df_3d_person.drop_duplicates('자치구', keep='first')['자치구']
df_person_count = df_3d_person['사람 수']

print(df_year)
year_list = list(df_year)
region_list = list(df_region)
print(year_list)
print(region_list)
```

```python
#3d bar 그리기 위해 비수치 데이터 수치화(인덱스로 변경함)
df_3d_person['년도_num']=0
df_3d_person['자치구_num']=1
df_3d_person
```

```python
#3d bar에 넣을 데이타 리스트 만들기
data_3d_person_list=[]
for i in range(len(df_3d_person)):
    data_row_list = list(df_3d_person.loc[i]) 
    #a = data_row_list[0]
    df_year_idx = list(df_year).index(data_row_list[0])
    df_region_idx = list(df_region).index(data_row_list[1])
    
    df_3d_person.loc[i,'년도_num'] = df_year_idx
    df_3d_person.loc[i,'자치구_num'] = df_region_idx
    
    data_3d_person_list.append([df_region_idx,df_year_idx,data_row_list[2]])
print(data_3d_person_list)
```

```python
data_3d_person_list
labels_3dp=['자치구','년도','인구수']
df_write = pd.DataFrame.from_records(data_3d_person_list,columns=labels_3dp)
df_write
```

```python
df_write.to_csv('for_3dPerson_data.csv',index=False)
```

```python
df_3dp_list=[]
for i in range(len(df_3dp_read)):
    df_3dp_list.append(list(df_3dp_read.loc[i]))

df_3dp_list
```

```python
# Tooltip callback 함수
# javascripthon을 사용하여 파이썬함수를 Javascript 함수로 변경한다.
#그래서 내부에 javascript 함수를 사용해도 괜찮다.
#콜백 함수이기 때문에 외부에서 데이터를 받아올 수 없다. 파라미터로 해결해야함. 그래서 index를 입력할 list는 직접 적어줌
def label_formatter(params):
    vals = params.value
    year_idx=vals[0]
    region_idx=vals[1]
    df_year_list = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
    df_region_list =['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']
    re_year = df_year_list[year_idx]
    re_region = df_region_list[region_idx]
    
    #result_str = '년도:'+str(re_year)+', 자치구:'+str(re_region)+', 유기견 수:'+vals[2]
    header_str = '''
                <div>
                  <span>구별 인구 수</span>
                  <br>
                '''
    result_str = header_str + params.marker + '<br>' + '- 년도 : '+re_year+'<br>'+'- 자치구 : '+re_region+' <br>'+'- 인구 수 : '+vals[2]+'명 <br></div>'
    return result_str
```

```python
# 구별, 년도별 인구 수
bar3d_person = Bar3D("", width=1200, height=600)

x_axis = year_list #년도 2005-2018
y_axis = region_list #자치 구


data = data_3d_person_list
range_color = ['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
               '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
bar3d_person.add(
    "구별, 년도별 인구 수",
    x_axis,
    y_axis,
    [[d[1], d[0], d[2]] for d in data],
    grid3d_width=100,
    grid3d_depth=250,
    
    datazoom_type='both',
    datazoom_orient='horizontal',
    
    #visualMap 설정
    is_visualmap=True,
    visual_range=[0, 600000],
    visual_range_color=range_color,
    visual_top='bottom',

    #x축 설정
    xaxis3d_name='년도',
    xaxis3d_interval=0,  #0으로 하면 모든 레이블 강제 표시 (axis3D 설정 봐야함)
    xaxis3d_margin=5,
    
    #y축 설정
    yaxis3d_name='자치구',
    yaxis3d_interval=0,
    
    #z축 설정
    zaxis3d_name='인구 수',
    
    #tooltip 설정
    tooltip_trigger='item',
    tooltip_formatter=label_formatter
)

#이건 전체 표출되는 화면 크기
bar3d_person.width=1000 #너비
bar3d_person.height=1000 #높이

bar3d_person
```

```python
#구별, 년도 별 인구 대비 유기견 수  (유기견/인구수)
df_3d_ps_dog = df.iloc[:,[0,1,2,5]]

#예쁜 시각화를 위해 자치구가 '합계'인 ROW 삭제
#연도 컬럼을 인덱싱 -  자치구 값이 '합계'인 ROW 삭제 후 ROW INDEX도 삭제되어 리셋하기 위함 
#년도로 인덱싱된 데이터 프레임의 인덱스를 정수로 리셋
df_3d_ps_dog = df_3d_ps_dog[df_3d_ps_dog['자치구'] != '합계'].set_index('년도').reset_index()

#3d bar 그리기 위해 비수치 데이터 수치화(인덱스로 변경함)
df_3d_ps_dog['년도_num']=0
df_3d_ps_dog['자치구_num']=1
df_3d_ps_dog['dog_per_person']= df_3d_ps_dog['유기견수']/df_3d_ps_dog['사람 수']

df_3d_ps_dog
```

```python
#축을 만들기 위해 연도, 자치구 리스트 뽑기
#연도, 지역 - 컬럼으로 추출
#X = df_year, Y = df_regiond, data = 인구 수
        
df_year = df_3d_ps_dog.drop_duplicates('년도', keep='first')['년도']
df_region = df_3d_ps_dog.drop_duplicates('자치구', keep='first')['자치구']
df_person_count = df_3d_ps_dog['사람 수']

year_list = list(df_year)
region_list = list(df_region)
print(year_list)
print(region_list)
```

```python
#3d bar에 넣을 데이타 리스트 만들기
df_3d_ps_dog_list=[]
for i in range(len(df_3d_ps_dog)):
    data_row_list = list(df_3d_ps_dog.loc[i]) 
    #print(data_row_list)
    #a = data_row_list[0]
    df_year_idx = list(df_year).index(data_row_list[0])
    df_region_idx = list(df_region).index(data_row_list[1])
    
    df_3d_ps_dog.loc[i,'년도_num'] = df_year_idx
    df_3d_ps_dog.loc[i,'자치구_num'] = df_region_idx
    
    df_3d_ps_dog_list.append([df_region_idx,df_year_idx,data_row_list[6]])
print(df_3d_ps_dog_list)
```

```python
# Tooltip callback 함수
# javascripthon을 사용하여 파이썬함수를 Javascript 함수로 변경한다.
#그래서 내부에 javascript 함수를 사용해도 괜찮다.
#콜백 함수이기 때문에 외부에서 데이터를 받아올 수 없다. 파라미터로 해결해야함. 그래서 index를 입력할 list는 직접 적어줌
def label_formatter_ps_dog2(params):
    vals = params.value
    year_idx=vals[0]
    region_idx=vals[1]
    df_year_list = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
    df_region_list =['종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']
    re_year = df_year_list[year_idx]
    re_region = df_region_list[region_idx]
    
    #result_str = '년도:'+str(re_year)+', 자치구:'+str(re_region)+', 유기견 수:'+vals[2]
    header_str = '''
                <div>
                  <span>인구대비 유기견 비율</span>
                  <br>
                '''
    result_str = header_str + params.marker + '<br>' + '- 년도 : '+re_year+'<br>'+'- 자치구 : '+re_region+' <br>'+'- 인구 대비 유기율 : '+(vals[2]*100)+'% <br></div>'
    return result_str
```

```python
# 구별, 년도별 인구 수
bar3d_ps_dog = Bar3D("", width=1200, height=600)

x_axis = year_list #년도 2005-2018
y_axis = region_list #자치 구


data = df_3d_ps_dog_list
range_color = ['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
               '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
bar3d_ps_dog.add(
    "구별, 년도별 인구 수",
    x_axis,
    y_axis,
    [[d[1], d[0], d[2]] for d in data],
    grid3d_width=100,
    grid3d_depth=250,
    
    datazoom_type='both',
    datazoom_orient='horizontal',
    
    #visualMap 설정
    is_visualmap=True,
    visual_range=[0, 0.003],
    visual_range_color=range_color,
    visual_top='bottom',

    #x축 설정
    xaxis3d_name='년도',
    xaxis3d_interval=0,  #0으로 하면 모든 레이블 강제 표시 (axis3D 설정 봐야함)
    xaxis3d_margin=5,
    
    #y축 설정
    yaxis3d_name='자치구',
    yaxis3d_interval=0,
    
    #z축 설정
    zaxis3d_name='인구 수',
    
    #tooltip 설정
    tooltip_trigger='item',
    tooltip_formatter=label_formatter_ps_dog2
)

#이건 전체 표출되는 화면 크기
bar3d_ps_dog.width=1000 #너비
bar3d_ps_dog.height=1000 #높이

bar3d_ps_dog
```
