```python
import pandas as pd
```

```python
#데이터 1 
df01 = pd.read_excel('데이터/01_서울시 주민등록인구 (구별)통계_05-18.xls', header=2)
df01_ft=df01[['기간', '자치구','계.1','남자.1','여자.1']]
df01_ft.rename(columns={'기간':'년도','계.1':'사람 수','남자.1':'남자수', '여자.1':'여자수'}, inplace = True)
df01_ft
```

```python
#데이터 2
df02= pd.read_excel('데이터/02_서울시 유기동물보호 현황 통계(05~18).xls', header=2)
df02_ft=df02[['기간','자치구','계','인도(주인)','입양분양','폐사안락사','계류기증']]
df02_ft.rename(columns={'기간':'년도',"계":"유기견수","인도(주인)":"복귀"}, inplace = True)
df02_ft
```

```python
#데이터 결과 = 데이터 1 + 데이터 2
df_result=pd.merge(df01_ft,df02_ft)
df_result
```

```python
#데이터 3
df03 = pd.read_excel('데이터/03_서울시 반려동물 유무 및 취득 경로 통계(14~18).xls', header=1)
df03.rename(columns={"대분류":"자치구", "기간":"년도"}, inplace = True)
df03
```

```python
#데이터 결과 = 데이터 결과 +데이터 3
df_result = pd.merge(df_result, df03, how='outer')
df_result
```

```python
#데이터 6
df06 = pd.read_excel('데이터/06_서울시 가구원수별 가구수(구별) 통계(05,10,15,16,17,18).xls', header=1)
df06.rename(columns={'구분':'자치구', '기간':'년도'}, inplace = True)
df06
```

```python
#데이터 결과 = 데이터 결과 + 데이터 6
df_result = pd.merge(df_result, df06, how='outer')
df_result
```
